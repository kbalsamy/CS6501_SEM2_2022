```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', 27)

url_dataset = "https://raw.githubusercontent.com/evassev/CS6501_SEM2_2022/main/Etivities/data/trade_combined.csv"
df = pd.read_csv(url_dataset)

df.head(10)
```

```python
df['Price'].plot.density()
```

```python
size = df.shape

for column in df.columns:
    # uniqueValues = df[column].unique()
    if len(df[column].unique()) ==size[0]:
        df.drop(column, axis=1, inplace=True)
        print(f"remove {column}")

df.head(5)
```

```python
unique_counts = df.nunique()
# print(unique_counts)
single_unique_columns = unique_counts[unique_counts == 1].index
# print(single_unique_columns)
df.drop(columns=single_unique_columns, axis=1, inplace=True)

df.shape
```

```python
for column in df.columns:
    df = df[df[column].notna()]

df.head(10)
```

```python
non_numeric_features = df.select_dtypes(exclude=np.number).columns.tolist()
print(non_numeric_features)
```

```python


date_features=["MaturityDate", "IssueDate", "SettlDate"]

for item in date_features:
    # print(item)
    df[item] = pd.to_datetime(df[item]).dt.strftime("%Y%m%d").astype(int)

df["TransactTime"] = pd.to_datetime(df["TransactTime"]).astype(int)

df.head(5)
```
